```python
origin_cell = 'A549'
origin_tf = 'ETS1'
origin_binding = 'AGCCGGAAGTT'
origin_cpg = 3

target_cell = 'GM12878' # GM12878 HepG2 K562 A549
target_tf = 'ETS1'
```

```python
origin_processed = f'processed/{origin_cell}_{origin_tf}_{origin_binding}.bed'

target_peaks = f'data/{target_cell}_{target_tf}_peaks.bed'
target_mehyl_plus = f'data/{target_cell}_wgbs_cpg_plus.bigWig'
target_mehyl_minus = f'data/{target_cell}_wgbs_cpg_minus.bigWig'
```

```python
# Load origin_processed

import pandas as pd

df_origin = pd.read_csv(origin_processed, sep='\t')
```

```python
df_origin
```

```python
# Binding_start_absolute

df_origin['binding_start_absolute'] = df_origin['1'] + df_origin['binding_start']
df_origin['binding_end_absolute'] = df_origin['1'] + df_origin['binding_end']
```

```python
# Load target peaks

df_target_peaks = pd.read_csv(target_peaks, sep='\t',header=None)
```

```python
df_target_peaks
```

```python
from intervaltree import Interval, IntervalTree

# Create an IntervalTree for the target peaks
tree = IntervalTree()
for idx, row in df_target_peaks.iterrows():
    tree.add(Interval(row[1], row[2]))

# Function to find overlapping regions
def find_overlaps(start, end):
    return tree[start:end]

# Apply the function to each row in df_origin
df_origin['overlaps_w_target'] = df_origin.apply(lambda row: find_overlaps(row['binding_start_absolute'], row['binding_end_absolute']), axis=1)

# Display the results
df_origin[['binding_start_absolute', 'binding_end_absolute', 'overlaps_w_target']]

```

```python
# Get methylation at target peaks

import pyBigWig

bw_plus = pyBigWig.open(target_mehyl_plus)   
bw_minus = pyBigWig.open(target_mehyl_minus)

def get_methylation_level(row):
    chrom = row[0]
    start = row['binding_start_absolute']
    end = row['binding_end_absolute']
    
    # Get the methylation level over the region
    methylation_plus = bw_plus.values(chrom, start, end)
    methylation_minus = bw_minus.values(chrom, start, end)
    
    return methylation_plus, methylation_minus

# Apply the function to get methylation levels
methylation_levels = df_origin.apply(get_methylation_level, axis=1)

# Assing
df_origin['target_methyl_plus'] = methylation_levels.apply(lambda x: x[0])
df_origin['target_methyl_minus'] = methylation_levels.apply(lambda x: x[1])
```

```python
# Classifiy the peaks according to no-methyl | c-methyl | g-methyl | cg-methyl at postiions 3and4

def classify_peak(row):

    # Get methylation values at positions 3 and 4 (0-indexed)
    val_c = row['target_methyl_plus'][origin_cpg]
    val_g = row['target_methyl_minus'][origin_cpg+1]

    # Consider a value > 0 (and not NaN) as methylated
    c_meth = pd.notnull(val_c) and val_c > 0
    g_meth = pd.notnull(val_g) and val_g > 0
    
    if not c_meth and not g_meth:
        return 'no-methyl'
    elif c_meth and not g_meth:
        return 'c-methyl'
    elif not c_meth and g_meth:
        return 'g-methyl'
    elif c_meth and g_meth:
        return 'cg-methyl'

df_origin['target_methylation_class'] = df_origin.apply(classify_peak, axis=1)
```

```python
# Check DNAse in target cell

df_target_dnase = pd.read_csv(f'data/{target_cell}_dnase.bed', sep='\t', header=None)
```

```python
df_target_dnase
```

```python
# Create an IntervalTree for the target DNase regions
dnase_tree = IntervalTree()
for idx, row in df_target_dnase.iterrows():
    dnase_tree.add(Interval(row[1], row[2]))

# Function to find overlapping DNase regions
def find_dnase_overlaps(start, end):
    return dnase_tree[start:end]

# Apply the function to each row in df_origin
df_origin['dnase_overlaps'] = df_origin.apply(lambda row: find_dnase_overlaps(row['binding_start_absolute'], row['binding_end_absolute']), axis=1)

# Display the results
df_origin[['binding_start_absolute', 'binding_end_absolute', 'dnase_overlaps']]
```
