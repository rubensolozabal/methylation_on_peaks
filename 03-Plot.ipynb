```python


origin_cell = 'A549' # GM12878 HepG2 K562 A549
origin_tf = 'ETS1'
origin_binding = 'AGCCGGAAGTT'
origin_cpg = 3

target_cell = 'K562'
target_tf = 'ETS1'
```

```python
import pandas as pd

path = f'processed/{origin_cell}_{origin_tf}_{origin_binding}_target_{target_cell}.bed'

# Load the BED file into a DataFrame
df = pd.read_csv(path, sep='\t')
```

```python
df.iloc[0].overlaps_w_target
```

```python
from intervaltree import Interval, IntervalTree
```

```python
# Convert overlaps_w_target to a set of intervals
df['overlaps_w_target'] = df['overlaps_w_target'].apply(lambda x: IntervalTree(eval(x)))
df['dnase_overlaps'] = df['dnase_overlaps'].apply(lambda x: IntervalTree(eval(x)))
```

```python
df
```

```python
# Filter the rows where dnase overlap is empty

df = df[df['dnase_overlaps'].apply(lambda x: len(x) > 0)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_bind = df[df['overlaps_w_target'].apply(lambda x: len(x) > 0)]
df_notBind = df[df['overlaps_w_target'].apply(lambda x: len(x) == 0)]


# Convert the classes to categorical with all desired categories so that the pivot table keeps all combinations
df_bind['methylation_class'] = pd.Categorical(df_bind['methylation_class'], categories=['c-methyl', 'g-methyl', 'cg-methyl', 'no-methyl'])
df_bind['target_methylation_class'] = pd.Categorical( df_bind['target_methylation_class'], categories=['c-methyl', 'g-methyl', 'cg-methyl', 'no-methyl'])

df_notBind['methylation_class'] = pd.Categorical(df_notBind['methylation_class'], categories=['c-methyl', 'g-methyl', 'cg-methyl', 'no-methyl'])
df_notBind['target_methylation_class'] = pd.Categorical( df_notBind['target_methylation_class'], categories=['c-methyl', 'g-methyl', 'cg-methyl', 'no-methyl'])

# Create a pivot table to count occurrences of each combination of methylation_class and target_methylation_class
pivot_table_bind = df_bind.pivot_table(index='methylation_class', columns='target_methylation_class', aggfunc='size', fill_value=0)
pivot_table_notBind = df_notBind.pivot_table(index='methylation_class', columns='target_methylation_class', aggfunc='size', fill_value=0)

# Create the heatmap
fig, ax = plt.subplots(1,2,figsize=(6, 3))
sns.heatmap(pivot_table_bind, annot=True, fmt='d', cmap='viridis',ax=ax[0], cbar=False)
ax[0].set_title(f'Origin Binds -> Target Binds')


sns.heatmap(pivot_table_notBind, annot=True, fmt='d', cmap='viridis',ax=ax[1], cbar=False)
ax[1].set_title(f'Origin Binds -> Target No Binds')

# X and Y labels target and origin cell
for i in range(2):
    ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=45, horizontalalignment='right')
    ax[i].set_yticklabels(ax[i].get_yticklabels(), rotation=0, horizontalalignment='right')

    # Set the labels
    ax[i].set_xlabel(f'Target {target_cell}')
    ax[i].set_ylabel(f'Origin {origin_cell}')

fig.subplots_adjust(wspace=1)

# No tight layout to avoid clipping the labels
plt.tight_layout()
```

```python
# Save pdf
fig.savefig(f'plots/{origin_cell}_{origin_tf}_{origin_binding}_target_{target_cell}_heatmap.pdf')
```
